```python
# Importing the packages needed for data visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Settig up the grid colour to white
sns.set_style('whitegrid')
```

```python
# Loading the data from the csv file
titanic = sns.load_dataset('titanic')
```

```python
# Jointplot

sns.jointplot(x='fare',y='age',data=titanic)
```

```python
# Here is the jointplot with for the variables fare Vs age . Here there is no corelation between these two variables.
```

```python
#distplot

sns.distplot(titanic['fare'],bins=30,kde=False,color='red')
```

```python
# There is no normal distribution with the data shown in the above graph
```

```python
#Boxplot

sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```
